The  rotation matrix ($R$) which rotates a 2D point around the origin with angle $\theta$  is given by:

$$R = \left[
	\begin{array}{r r}
		\cos(\theta) & -\sin(\theta)\\
		\sin(\theta) & \cos(\theta)
	\end{array}
\right],$$

1. **Create rotation matrices:** Implement the function `get_2d_rotation_matrix()`
 below which, given an input angle
`theta`
 (in radians) returns the rotation matrix `R`
:



```python
def get_2d_rotation_matrix(theta):
    """Return a rotation matrix for the specified input angle theta."""
    # PLACEHOLDER <<
    # ...
    # >>
    # SOLUTION <<
    return np.array([[ math.cos(theta), -math.sin(theta) ],
                     [ math.sin(theta),  math.cos(theta) ]], dtype=np.float64)
    # >>
```
