# Pen and paper exercises
These exercises should be completed using only pen and paper. The goal 
is to test some of the basic linear algebra skills needed for proper 
understanding of the course material.

**Task 1 (medium)♾️**
Determine if the following statements are true or false. If a statement is true, provide an argument. If a statement is 
false, provide a counterexample:
1. If $A^2$ is defined then $A$ is necessarily square.
2. If $AB$ and $BA$ are defined then $A$ and $B$ are square.
3. If $AB$ and $BA$ are defined then $AB$ and $BA$ are square.
4. If $AB=B$ then $A=I$.
