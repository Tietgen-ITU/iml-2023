Slicing allows selecting a sequence of array elements by using the `<start>:<stop>`

## Using Numpy for array operations
While these implementations seem fine for our sample inputs, they quickly become unbearingly slow when increasing the
size of the operands.
Let‚Äôs try an example. The code below uses numpy to generate $1000000$-dimensional vectors of random numbers:


````python
ta = np.random.randint(100, size=1000000)
tb = np.random.randint(100, size=1000000)
````

Jupyter Lab provides special command `%timeit <statement>`
 which runs a performance test on the supplied statement. This 
way, we can test how fast your native implementation of the inner product is:


````python
%timeit dot(ta, tb)
````

Not very fast huh? Now let‚Äôs try using numpy‚Äôs built in function for inner products, `np.dot`
:


````python
%timeit np.dot(ta, tb)
````

That is approximately 300 times faster than the native version (on the test computer, anyway)!. What about other list operations? Lets try the sum function:


````python
%timeit sum(ta)
````

````python
%timeit np.sum(ta)
````

A similar performance increase. Because of its speed, Numpy should always be used instead of native Python wherever possible. In general, you should expect a speedup of one to three orders of magnitude when compared to native python.
